#### Задача 1. 

Дано некоторое многозначное число. Найдите сумму его цифр.

````python
def summa(number):
    return sum(int(i) for i in str(number))

number = int(input("чиселко: "))
print(f"сумма цифр: {summa(number)}")
````

#### Задача 2.

Процентная ставка по вкладу составляет P процентов годовых, которые прибавляются к сумме вклада. Вклад составляет X рублей Y копеек. Определите размер вклада через год.
Программа получает на вход целые числа P, X, Y и должна вывести два числа: величину вклада через год в рублях и копейках. Дробная часть копеек отбрасывается.

````python
def deposit(p, x, y):
    total_cents = x * 100 + y
    total_cents += total_cents * p / 100
    rubli = int(total_cents // 100)
    kopeiki = int(total_cents % 100)
    return rubli, kopeiki

p = int(input("процент: "))
x = int(input("рубли: "))
y = int(input("копейки: "))
rubli, kopeiki = deposit(p, x, y)
print(f"размер вклада через год: {rubli} руб. {kopeiki} коп.")
````

#### Задача 3.

Вас назначили шпионом, и теперь ваша задача — передавать секретные сообщения. Но вот беда: ваши сообщения могут перехватить! Чтобы защитить их, нужно зашифровать текст. Вам предстоит написать программу, которая будет зашифровывать сообщение следующим образом:

1. Программа должна попросить пользователя ввести сообщение (строку).
2. Затем она должна пройтись по каждому символу строки и заменить:
   - каждую букву 'а' на 'о',
   - каждую букву 'е' на 'и',
   - каждую букву 'о' на 'а',
   - каждую букву 'и' на 'е'.
3. Все остальные символы должны остаться без изменений.
4. В результате программа должна вывести зашифрованное сообщение.

Пример работы программы:

    Введите сообщение: привет, как дела?
    Зашифрованное сообщение: преавот, кок дола?

````python
#я не очень понимаю, как из "привет, как дела?" получить "преавот, кок дола?". там разве в таком случае не должно быть введено "приоват, как дало?"

def shpion(message):
    secret_message = ""
    for char in message:
        if char == 'а':
            secret_message += 'о'
        elif char == 'е':
            secret_message += 'и'
        elif char == 'о':
            secret_message += 'а'
        elif char == 'и':
            secret_message += 'е'
        else:
            secret_message += char
    return secret_message

message = input("введите сообщение: ")
print(f"зашифрованное сообщение: {shpion(message)}")
````

1. Программа должна запросить у пользователя число n, которое задает количество полосок в зебре.
2. Затем программа должна вывести на экран чередующиеся строки: первая строка состоит из символов "####", а вторая — из символов "....".
3. Например, если пользователь введет число n = 5, программа должна вывести:
  

        ####
        ....
        ####
        ....
        ####
   
4. Если пользователь введет четное число, например n = 4, программа должна начать с "####" и закончить на "....". Если число нечетное, программа должна закончить на "####".

Пример работы программы:


    Введите количество полосок: 5

    ####
    ....
    ####
    ....
    ####


````python
def poloski(n):
    for i in range(n):
        if i % 2 == 0:
            print("####")
        else:
            print("....")

n = int(input("количество полосок: "))
poloski(n)
````

#### Задача 5. 

Представьте, что вы разрабатываете калькулятор, который умеет выполнять только деление. Но иногда пользователи могут допустить ошибку и попытаться разделить на ноль, что, как известно, невозможно. Ваша задача — написать программу, которая будет аккуратно обрабатывать такие ошибки.

1. Программа должна запросить у пользователя два числа: делимое и делитель.
2. Затем она должна попытаться выполнить деление и вывести результат.
3. Если пользователь ввел делитель равный нулю, программа должна обработать эту ошибку и вывести сообщение: "Ошибка: Деление на ноль невозможно!"
4. Если пользователь ввел не число, программа должна вывести сообщение: "Ошибка: Введите числовое значение!"
5. Программа должна продолжать работать и снова запрашивать числа, пока пользователь не введет команду "выход".

Пример работы программы:

```
Введите делимое: 10
Введите делитель: 0
Ошибка: Деление на ноль невозможно!

Введите делимое: 10
Введите делитель: 2
Результат: 5.0

Введите делимое: abc
Ошибка: Введите числовое значение!

Введите делимое: выход
Программа завершена.
```

````python
#как будто работает медленно? работало ок днем, а ночью тупит. или это я туплю? 

def calculator():
    while True:
        try:
            input_str = input("введите делимое и делитель или 'выход' для завершения: ")
            if input("введите делимое и делитель или 'выход' для завершения: ").lower() == "выход":
                break
            dividend, divisor = map(float, input_str.split())
            if divisor == 0:
                raise ZeroDivisionError
            print(f"результат: {dividend / divisor}")
        except ValueError:
            print("ошибка: введите числовое значение!")
        except ZeroDivisionError:
            print("ошибка: деление на ноль невозможно!")

calculator()
````

#### Задача 6.

Числа Фибоначчи (простые). Вводится целое число: с помощью формулы выведите число Фибоначчи с этим номером. 

````python
import math

def fibonacci(n):
    phi = (1 + math.sqrt(5)) / 2
    psi = (1 - math.sqrt(5)) / 2
    return int((phi ** n - psi ** n) / math.sqrt(5))

n = int(input("номер числа: "))
print(f"чиселко: {fibonacci(n)}")
````

````python
# фибоначчи тяжелая с семинара
# я читала про мемоизацию несколько часов но не факт что поняла что-то больше чем то что она уменьшает количество повторных вычислений

def fibonacci(n, memo={}):                                      #делаю пустой словарь для мемоизации
    if n in memo:                                               #если оно уже вычислено, словарь его выдает
        return memo[n]
    if n == 0:                                                  #если 0 то 0))) (это я в основном для себя комменчу, комменты для вас обычно выше решения)
        return 0
    if n == 1:                                                  #если 1 то 1
        return 1
    memo[n] = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)   #если нет в мемо, считаем два предыдущих и складываем
    return memo[n]

n = int(input("номер числа: "))
print(f"чиселко: {fibonacci(n)}")
````

#### Задача 7.

Напишите программу, которая на вход принимает число секунд и возвращает его в часах, минутах и секундах в формате HH:MM:SS. 

````python
def time_sec(seconds):
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    seconds = seconds % 60
    return f"{hours:02}:{minutes:02}:{seconds:02}"

seconds = int(input("введите секунды: "))
print(f"время в формате HH:MM:SS: {time_sec(seconds)}")
````
